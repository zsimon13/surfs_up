# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()

```

```python
# create references to the tables
Measurement = Base.classes.measurement
Station = Base.classes.station

#create a session link to the database
session = Session(engine)
```

```python
# set up parametewrs to import rain data for the year
# use dt.date() and dt.timedelta() to set up one year span
prev_year = dt.date(2017, 8, 23) - dt.timedelta(days=365)

#retrieve precipitation "scores"
# set up variable for results adn add in session query
results = session.query(Measurement.date, Measurement.prcp)

#print(results.all())

#filter out data that is older than a year old
#use .all() to add all info from query and turn into a list
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= prev_year).all()

print(results)
```

```python
# save results into a DataFrame
df = pd.DataFrame(results, columns=['date', 'precipitation'])
```

```python
# set index of df to date column
df.set_index(df['date'], inplace=True)
print(df)
```

```python
# reprint the df without the index so date does not appear twice
print(df.to_string(index=False))
```

```python
# sort values by date to see weather trend
df=df.sort_index()
print(df.to_string(index=False))
```

```python
# make a plot to visualize the dataframe
df.plot()
plt.xticks(rotation = 90)
```

```python
# us .describe() to retrieve statistical data from the dataframe
df.describe()
```
